```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Constants
fs = 512  # Sampling frequency in Hz
n_samples = 100000  # Number of samples
duration = n_samples / fs  # Total duration for 10000 samples
f1 = 0.5  # Frequency of first wave (0.5 Hz)
f2 = 40  # Frequency of second wave (40 Hz)

# Time vector
t = np.arange(0, n_samples) / fs

# Generate the 1Hz and 40Hz sine waves
wave_1Hz = np.sin(2 * np.pi * f1 * t)
wave_40Hz = np.sin(2 * np.pi * f2 * t)

# Apply scaling to each wave
scaled_wave_1Hz = wave_1Hz
scaled_wave_40Hz = wave_40Hz / 20

# Combine the waves (add them together)
final_waveform = scaled_wave_1Hz + scaled_wave_40Hz

# Create a DataFrame
df = pd.DataFrame(final_waveform, columns=["Waveform"])

# Save to CSV
df.to_csv("test_waveform.csv", index=False)

print("CSV file 'test_waveform.csv' created with 10,000 entries.")
```

```python
# Plot the first n samples
n_visualize = 1000
plt.plot(t[:n_visualize], final_waveform[:n_visualize])
plt.title(f'First {n_visualize} samples of the waveform')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```
